```python
def clean_text(sentence):
  sentence = sentence.lower()
  #remove punctuations
  sentence = re.sub("[^a-z]+"," ",sentence) #anything which is not alphabat substiute with space

  sentence = sentence.split()
  sentence = [s for s in sentence if len(s)>1]
  sentence = " ".join(sentence)
  return sentence
```

```python
def preprocess_img(img):
  img = image.load_img(img,target_size=(224,224))
  img = image.img_to_array(img)
  #expand_dims or reshape can be used so that img_size (batch_size,224,224,3) cause that is the input_shape expected by resnt
  img = np.expand_dims(img,axis=0)
  #now img_size = (1,224,224,3)
  #normalisation - now each network has some specific type of preprocessing so for resnet we have imported preprocess_input
  #to serve that purpose
  img = preprocess_input(img)
  return img
```

```python
# method to encode img into image_features of shape 2048
def encode_img(img):
  img = preprocess_img(img)
  feature_vector = model_new.predict(img)
  #feature_vector's shape = 1,2048 so convert back into (2048,)
  feature_vector = feature_vector.reshape((-1,))
  #print(feature_vector.shape)
  return feature_vector
```

```python
word_to_index = {}
index_to_word = {}

#we reserve 0 index for ending the caption for eg we will feed sentence like - dog is running
# so word-to-index would be [12,45,2,0] dog's index = 12 and zero index for termination.

for i,word in enumerate(total_words):
  word_to_index[word] = i+1
  index_to_word[i+1] = word
```

```python
#add two special words
index_to_word[1846] = 'startseq'
word_to_index['startseq'] = 1846

index_to_word[1847] = 'endseq'
word_to_index['endseq'] = 1847

vocab_size = len(word_to_index) + 1
print("Vocab size: %d"%(vocab_size))
```

```python
#maxlength of the caption
max_len = 0
for key in train_descriptions.keys():
  for cap in train_descriptions[key]:
    max_len = max(max_len,len(cap.split()))


print(max_len)
```

```python
# How to genrate sentence ,so basiclly we feed starting tag to the modelor rnn cell then it will the probability of all 
#words and finds out the max prob index and then index to word adn then that wrd is feeded to the next rnn cell and it will
#predict next word .P(dog|<s>) , P(is|<s>,dog) , P(running|<s>,dog,is)
# so the prob of next word given that other previous words so this is known as Language Modelling

#for eg in markov rule of text gen, we look back upto 4 words but in rnn/lstm , we look back upto the entire sentence
#and then predict next word
#how lang modelling (recording)
```

```python
# Data Loader(Generator) - we will implement our own data gen so that batch of images+caption can be feeded to the network

def data_generator(train_descriptions,encoding_train,word_to_index,max_len,batch_size):
  X1,X2,y = [],[],[]
  #X1 - image features , X2 - captions and y the predicted
  n = 0
  while True:
    for key,caption_list in train_descriptions.items():
      n += 1
      photo = encoding_train[key]

      for caption in caption_list:
        seq = [word_to_index[word] for word in caption.split() if word in word_to_index]
        #ignoring unknown words
        for i in range(1,len(seq)):
          xi = seq[0:i]
          yi = seq[i]

          #padding 0 at the end of each seq as maxlen=35
          #o denote padding word thats why 0 is reserved
          xi = pad_sequences([xi],maxlen=max_len,value=0,padding='post')[0]
          yi = to_categorical([yi],num_classes=vocab_size)[0] #one-hot vector

          X1.append(photo)
          X2.append(xi)
          y.append(yi)

        if n == batch_size:
          yield [[np.array(X1),np.array(X2)],np.array(y)]
          X1 ,X2,y = [],[],[]  
          n = 0
```

```python
# Word Embeddings
# we will use TL - glove embeddings to convert caption word indices to glove vector embeddings
# so we need to construct embedding matrix of vocab words with dimension as vocab word size X 50 
# we will not be needing all 6billion words of glove6B50D.txt

!kaggle datasets download -d watts2/glove6b50dtxt 
```

```python
f = open("/content/glove.6B.50d.txt",encoding='utf8')
```

```python
embedding_index = {}
for line in f:
  values = line.split()
  #print(values)
  word = values[0]
  word_embedding = np.array(values[1:],dtype='float')
  embedding_index[word] = word_embedding
```

```python
embedding_index['dog']
```

```python
def get_embedding_matrix():
  emb_dim = 50
  matrix = np.zeros((vocab_size,emb_dim))
  for word,index in word_to_index.items():
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
      matrix[index] =  embedding_vector
    
  return matrix
```

```python
embedding_matrix = get_embedding_matrix()
embedding_matrix.shape
```

```python
# glove embeddings for the startseq and endseq are zero
embedding_matrix[1847]
```

```python
# Model Architecture
#follow medium article
# for image part we will 2048 dimensional encoded image features and that will be squeezed to 256 dimensional by model
# and for text part- partial seq or caption , we will have emebedding matrix that is feeded to rnn layer and we got a 256 dim vector
# those 256 text part and 256 image aprt are concatenated to single one and feeded to the mlp with last layer as size = vocab_Size
# and activation as softmax so it will give prob dist over entire voacb to predict the next word
#loss function would be categorical cross entropy as there are more than two classes

# since we know that input will be X1,X2 = image features and partial seq and y - one-hot vector of size= vocab_size for
#thee next word.so we will keras functional api 
```

```python
# Model Implementation
# first part to convert image feature vector into  smaller vector
# resnet50 model output (stored in a file,we are going to reda the features and fed)feeded to the input layer,
# then adding dropout -dropping some of the features randomly

input_img_features = Input(shape=(2048,))
inp_img1 = Dropout(0.3)(input_img_features)
inp_img2 = Dense(256,activation='relu')(inp_img1)
```

```python
#Captions as input
input_captions = Input(shape=(max_len,)) # each caption of maxlen =35 ,specify the maxlen of single sample,otherwise the input would be batch_sizeXmax_Len
inp_cap1 = Embedding(input_dim=vocab_size,output_dim=50,mask_zero=True)(input_captions)
inp_cap2 = Dropout(0.3)(inp_cap1)
inp_cap3 = LSTM(256)(inp_cap2) #256 is the size of hidden activation vector that comes out of the layer ,it is not no of cells in the layer
```

```python
# till now we have build two mini models one for preprocessing text other for preprocessing images
decoder1 = add([inp_img2,inp_cap3])  #concatenate,add takes tensors as input and concatenates them
decoder2 = Dense(256,activation='relu')(decoder1)
outputs = Dense(vocab_size,activation='softmax')(decoder2)


#combined model
model = Model(inputs=[input_img_features,input_captions],outputs=outputs)
model.summary()
```

```python
# now we have glove embedding so we need to set the wts of embeeding layer accordingly and then set it to nontrainble
model.layers[2].set_weights([embedding_matrix])
model.layers[2].trainable = False

model.compile(loss='categorical_crossentropy',optimizer='adam')
```

```python
# Training of Model

epochs = 20
batch_size = 3
steps = len(train_descriptions)//batch_size
print(len(train_descriptions))
```

```python
!mkdir model_weights
```

```python
def train_model():
  for i in range(epochs):
    generator = data_generator(train_descriptions,encoding_train,word_to_index,max_len,batch_size)
    model.fit_generator(generator,epochs=1,steps_per_epoch=steps,verbose=1)#epochs = how many epochs you want to train that batch
    #steps_per_epoch = how many steps it will take to passs the whole training set(for eg if there were 90 training egs 
    # and batch_size be 3 then 30 will be the no of steps)
    model.save("./model_weights/model_"+str(i)+".h5")
```

```python
#Prediction 

def predict_caption(photo):
  in_text = 'startseq'
  for i in range(max_len):
    sequence = [word_to_index[w] for w in in_text.split() if w in word_to_index]
    sequence = pad_sequences([sequence],maxlen = max_len,padding='post')

    y_pred = model.predict([photo,sequence])
    y_pred = y_pred.argmax() #we can pick top three or four words 
    #but here we are picking with the max prob - greedy sampling
    word = index_to_word[y_pred]
    in_text += (' ' + word)

    if word == 'endseq':
      break

  final_caption = in_text.split()[1:-1] #remove startseq and endseq
  final_caption = ' '.join(final_caption) #joining the words again
  return final_caption
```
